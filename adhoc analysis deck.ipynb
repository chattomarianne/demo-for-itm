<font size="5">Getting the number of Repeaters, Inactive, and Engaged Customers

<font size="4"> Define repeaters, inactive, engaged
- Repeaters : 1 if customer of the current month had also a transaction from the previous month
- Inactive  : 1 if customer of the current month DID not have transaction but had previous transaction the past months
- Engaged   : 1 if customer of the current month HAD transaction from the previous months(consistent)


<font size="4"> Create a Table for Transaction History

```python
def to_user_transaction_history_df(trdf):
    user_history_dict = dict()
    trans_month = final_df.month.drop_duplicates().to_list()
    for month in trans_month:
        a = trdf[(trdf.month == month)]
        b = a.groupby(['mail'])['transaction_id'].count()
        b = b.to_dict()
        user_history_dict.update({month: b})    
    return pd.DataFrame(user_history_dict)

user_history_df = to_user_transaction_history_df(trdf)
user_history_df
```
